### 1. Crosstab analysis

### 2. Statistical Tests 

```python
from scipy.stats import chi2_contingency

# Chi-square test for 'category' and 'isBestSeller'
ct_chi = pd.crosstab(df['category'], df['isBestSeller']).sort_values(by = True, ascending = False)
chi2_statistic, chi2_p_value, _, _ = chi2_contingency(ct_chi)

chi2_statistic, chi2_p_value
```

```python
# Perform Cramér's V test
from scipy.stats.contingency import association
association(ct_chi, method="cramer")
```

## Part 2

### 0. Remove outliers

```python
def remove_outliers(data, column):
    # Calculate Q1 and Q3 for the specified column
    Q1 = data[column].quantile(0.25)
    Q3 = data[column].quantile(0.75)
    IQR = Q3 - Q1

    # Define bounds for the outliers
    lower_bound = Q1 - 1.5 * IQR
    upper_bound = Q3 + 1.5 * IQR

    # Identify the outliers in the specified column
    outliers = data[(data[column] < lower_bound) | (data[column] > upper_bound)]

    # Drop rows that have outliers in the specified column
    new_data = data.drop(index=outliers.index)
    
    return new_data
```

```python
df_out = remove_outliers(df, 'price')
```
